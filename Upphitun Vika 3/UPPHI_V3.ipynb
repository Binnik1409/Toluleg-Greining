```python
import numpy as np

def create_matrix_A(n, alpha, beta):
    'Creates a specific n x n matrix with given alpha value.'

    I = np.ones(n)
    b = I * beta
    I = np.ones(n-1)
    a = I * alpha
    matrix = np.diag(b) + np.diag(a,-1) + np.diag(a,1)
    matrix[0][0] = 1
    matrix[0][1] = 0
    matrix[n-1][n-2] = 0
    matrix[n-1][n-1] = 1

    return matrix
```

```python
import functions as f
import numpy as np
import matplotlib.pyplot as plt

y0 = 1
yn = -5
n = 20
L = np.pi/2
h = L/(n-1)

x = np.linspace(0, np.pi/2, n)

alpha = 1
beta = -2 + h**2

matrix_a = f.create_matrix_A(n, alpha, beta)
vector_b = np.zeros(n)
vector_b[0] = y0
vector_b[-1] = yn

sol = np.linalg.solve(matrix_a, vector_b)

ref_sol = np.cos(x) - 5 * np.sin(x)

plt.plot(x, sol, marker='o', label='Numerical Solution')
plt.plot(x, ref_sol, label='Reference Solution', linestyle='--')
plt.title('Numerical Solution of BVP')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.show()
```
